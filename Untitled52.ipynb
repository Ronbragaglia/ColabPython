```python
from google.colab import drive
drive.mount('/content/drive')
PATH = '/content/drive/MyDrive'
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plot
import glob

def alterar_gama(image, gamma):
    invGamma = 1.0 / gamma
    table = np.array([((i / 255.0) ** invGamma) * 255
        for i in np.arange(0, 256)]).astype("uint8")
    image = cv.LUT(image, table)
    return image

def corrigir_contraste(image):
    histogram = cv.hist(image)

    return image

def converter_cinza(image):
    image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
    return image

def binarizar_imagem(image):
    return cv.threshold(image, 190, 255, cv.THRESH_OTSU)[1]

def aplicar_blur(image):
    return cv.blur(image, (5, 5))

def extrair_medidas(img_bin):
    medidas = []
    (numCCs, imgLabels, stats, centroids) = cv.connectedComponentsWithStats(img_bin, connectivity=8)
    for i in range(1, len(stats)):
        top_c, top_l, larg, alt, area = stats[i, [cv.CC_STAT_LEFT, cv.CC_STAT_TOP, cv.CC_STAT_WIDTH, cv.CC_STAT_HEIGHT, cv.CC_STAT_AREA]]
        medidas.append([larg, alt, area])
    return medidas

def contagem(medidas):
    # Largura = 0 - Altura = 1 - Area = 2
    parafusos = 0
    porca = 0
    for medida in medidas:
        if medida[2] > 20:
            if 10 <= medida[0] <= 40 and 10 <= medida[1] <= 40:
                porca += 1
            else:
                parafusos += 1
    return parafusos, porca

def contador_parafusos(image_path, print_medidas=False, print_image=False, testing=False):
    print("Resultado imagem:" + image_path.split("/")[-1])
    image = cv.imread(image_path)
    image = alterar_gama(image, 0.31)
    image = converter_cinza(image)
    image = aplicar_blur(image)
    image_bin = binarizar_imagem(image)
    if print_image:
        plot.imshow(image_bin, cmap='gray')
        plot.savefig(image_path.replace(".jpg", "_bin.jpg"))
        plot.show()
    medidas = extrair_medidas(image_bin)
    if print_medidas:
        print(medidas)
    resultado = contagem(medidas)
    parafusos = resultado[0]
    porcas = resultado[1]
    print("-------------------------")
    if porcas == parafusos and porcas + parafusos == 20:
        print("Aprovado.")
    else:
        print("Reprovado.")
        print(f"Falta(m) {10 - parafusos} parafusos e {10 - porcas} porcas.")
    print(f"Parafusos: {parafusos}, Porcas: {porcas}")
    print("-------------------------")
    print("-------------------------")
    if testing:
        return [parafusos, porcas]

p = "/content/drive/MyDrive/EP1_Esp_IA_AM_2024/Imagens/img05.jpg"
contador_parafusos(p, False, True, True)

massa_teste = [
    ["/content/drive/MyDrive/EP1_Esp_IA_AM_2024/Imagens/img01.jpg", "10,10"],
    ["/content/drive/MyDrive/EP1_Esp_IA_AM_2024/Imagens/img02.jpg", "10,10"],
    ["/content/drive/MyDrive/EP1_Esp_IA_AM_2024/Imagens/img03.jpg", "10,10"],
    ["/content/drive/MyDrive/EP1_Esp_IA_AM_2024/Imagens/img04.jpg", "10,10"],
    ["/content/drive/MyDrive/EP1_Esp_IA_AM_2024/Imagens/img05.jpg", "9,9"],
    ["/content/drive/MyDrive/EP1_Esp_IA_AM_2024/Imagens/img06.jpg", "10,10"],
    ["/content/drive/MyDrive/EP1_Esp_IA_AM_2024/Imagens/img07.jpg", "7,10"],
    ["/content/drive/MyDrive/EP1_Esp_IA_AM_2024/Imagens/img08.jpg", "7,9"],
    ["/content/drive/MyDrive/EP1_Esp_IA_AM_2024/Imagens/img09.jpg", "10,10"],
    ["/content/drive/MyDrive/EP1_Esp_IA_AM_2024/Imagens/img10.jpg", "10,9"],
    ["/content/drive/MyDrive/EP1_Esp_IA_AM_2024/Imagens/img11.jpg", "10,0"],
    ["/content/drive/MyDrive/EP1_Esp_IA_AM_2024/Imagens/img12.jpg", "10,10"],
    ["/content/drive/MyDrive/EP1_Esp_IA_AM_2024/Imagens/img13.jpg", "0,10"],
    ["/content/drive/MyDrive/EP1_Esp_IA_AM_2024/Imagens/img14.jpg", "7,7"],
    ["/content/drive/MyDrive/EP1_Esp_IA_AM_2024/Imagens/img15.jpg", "9,10"]
]

def testar_massa(test_list):
    corretas = 0
    for test in test_list:
        resultado = contador_parafusos(test[0], testing=True)
        parafusos = resultado[0]
        porcas = resultado[1]
        if f"{parafusos},{porcas}" == test[1]:
            print(f"Imagem {test[0]} foi interpretada corretamente, encontrou: {parafusos} parafusos e {porcas} porcas")
            corretas += 1
        else:
            print(f"ATENÇÃO!! Imagem {test[0]} não foi interpretada corretamente, encontrou: {parafusos} parafusos e {porcas} porcas.")
            print(f"O correto seria: {str(test[1])}")
    print(f"O teste foi concluído com {(100 / len(test_list)) * corretas}% de precisão.")

testar_massa(massa_teste)

def run(folder_path):
    if folder_path[len(folder_path) - 1] != "/":
        folder_path = folder_path + "/"
    folder_path = folder_path + "*"
    files = glob.glob(folder_path)
    for f in files:
        try:
            contador_parafusos(f)
        except:
            print(f"Erro tentar processar a imagem no caminho: {f}")

pasta = "/content/drive/MyDrive/EP1_Esp_IA_AM_2024/Imagens/"
run(pasta)
```
