```python
import requests
import pandas as pd

url = "https://pt.wikipedia.org/wiki/Premier_League_de_2023%E2%80%9324"


requisicao = requests.get(url)
if requisicao.status_code == 200:

    INDEX_TABELA_CLASSIFICACAO = 7
    INDEX_TABELA_JOGOS = 10


    tabelas = pd.read_html(requisicao.text)


    if len(tabelas) > INDEX_TABELA_CLASSIFICACAO and len(tabelas) > INDEX_TABELA_JOGOS:
        tabela_classificacao = tabelas[INDEX_TABELA_CLASSIFICACAO]
        tabela_jogos = tabelas[INDEX_TABELA_JOGOS]


        display(tabela_classificacao)
        display(tabela_jogos)
    else:
        print("Índices de tabela fora do intervalo.")
else:
    print(f"Falha na requisição. Código de status: {requisicao.status_code}")
```
