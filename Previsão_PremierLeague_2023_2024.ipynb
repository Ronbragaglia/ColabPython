```python
import pandas as pd
import requests
```

```python
requisicao = requests.get("https://pt.wikipedia.org/wiki/Premier_League_de_2023%E2%80%9324")

tabelas = pd.read_html(requisicao.text)
tabela_classificacao = tabelas[7]
tabela_jogos = tabelas[10]
display(tabela_classificacao)
display(tabela_jogos)
```

```python
nomes_times = list(tabela_jogos["Casa \ Fora"])
apelidos = list(tabela_jogos.columns)
apelidos.pop(0)
print(nomes_times)
print(apelidos)

de_para_times = dict(zip(apelidos, nomes_times))
print(de_para_times)
```

```python
tabela_jogos_ajustada = tabela_jogos.set_index("Casa \ Fora")
tabela_jogos_ajustada = tabela_jogos_ajustada.unstack().reset_index()
tabela_jogos_ajustada = tabela_jogos_ajustada.rename(columns={"level_0": "fora", "Casa \ Fora": "casa", 0: "resultado"})

def ajustar_apelido_time(linha):
    apelido = linha["fora"]
    nome = de_para_times[apelido]
    return nome

tabela_jogos_ajustada["fora"] = tabela_jogos_ajustada.apply(ajustar_apelido_time, axis=1)
tabela_jogos_ajustada = tabela_jogos_ajustada[tabela_jogos_ajustada["fora"]!=tabela_jogos_ajustada["casa"]]
# pd.set_option('display.max_rows', None)
display(tabela_jogos_ajustada)
```

```python
tabela_jogos_ajustada["resultado"] = tabela_jogos_ajustada["resultado"].fillna("a jogar")
tabela_jogos_realizados = tabela_jogos_ajustada[tabela_jogos_ajustada["resultado"].str.contains("–")]
display(tabela_jogos_realizados)
tabela_jogos_faltantes = tabela_jogos_ajustada[~tabela_jogos_ajustada["resultado"].str.contains("–")]
tabela_jogos_faltantes = tabela_jogos_faltantes.drop(columns=["resultado"])
# pd.set_option('display.max_rows', None)
display(tabela_jogos_faltantes)
```

```python
tabela_jogos_realizados[["gols_casa", "gols_fora"]] = tabela_jogos_realizados["resultado"].str.split("–", expand=True)
tabela_jogos_realizados = tabela_jogos_realizados.drop(columns=["resultado"])
tabela_jogos_realizados["gols_casa"] = tabela_jogos_realizados["gols_casa"].astype(int)
tabela_jogos_realizados["gols_fora"] = tabela_jogos_realizados["gols_fora"].astype(int)
# pd.set_option('display.max_rows', None)
display(tabela_jogos_realizados)
```

```python
# media gols feitos dentro de casa
# media gols sofridos dentro de casa
media_gols_casa = tabela_jogos_realizados.groupby("casa").mean(numeric_only=True)
media_gols_casa = media_gols_casa.rename(columns={"gols_casa": "gols_feitos_casa", "gols_fora": "gols_sofridos_casa"})
display(media_gols_casa)

# media gols feitos fora de casa
# media gols sofridos fora de casa
media_gols_fora = tabela_jogos_realizados.groupby("fora").mean(numeric_only=True)
media_gols_fora = media_gols_fora.rename(columns={"gols_casa": "gols_sofridos_fora", "gols_fora": "gols_feitos_fora"})
display(media_gols_fora)
```

```python
tabela_estatisticas = media_gols_casa.merge(media_gols_fora, left_index=True, right_index=True)
tabela_estatisticas = tabela_estatisticas.reset_index()
tabela_estatisticas = tabela_estatisticas.rename(columns={"casa": "time"})
display(tabela_estatisticas)
```

```python
# Teste de hipótese para comparar as médias de gols feitos em casa e fora de casa para cada time
from scipy.stats import ttest_rel

p_values_casa_fora = []
for _, row in tabela_estatisticas.iterrows():
    time = row["time"]
    gols_casa = tabela_jogos_realizados[tabela_jogos_realizados["casa"] == time]["gols_casa"]
    gols_fora = tabela_jogos_realizados[tabela_jogos_realizados["fora"] == time]["gols_fora"]
    min_len = min(len(gols_casa), len(gols_fora))
    gols_casa = gols_casa[:min_len]
    gols_fora = gols_fora[:min_len]
    p_value = ttest_rel(gols_casa, gols_fora).pvalue
    p_values_casa_fora.append(p_value)
tabela_estatisticas["p_value_casa_fora"] = p_values_casa_fora

# Exibindo a tabela de estatísticas com os p-values
display(tabela_estatisticas)

```

```python
from scipy.stats import f_oneway

# Lista para armazenar os p-values da ANOVA
p_values_anova = []

# Iterando sobre cada time na tabela de estatísticas
for _, row in tabela_estatisticas.iterrows():
    time = row["time"]

    # Obtendo os gols feitos em casa e fora de casa para o time atual
    gols_casa = tabela_jogos_realizados[tabela_jogos_realizados["casa"] == time]["gols_casa"]
    gols_fora = tabela_jogos_realizados[tabela_jogos_realizados["fora"] == time]["gols_fora"]

    # Calculando o p-value da ANOVA para os gols feitos em casa e fora de casa
    p_value_anova = f_oneway(gols_casa, gols_fora).pvalue
    p_values_anova.append(p_value_anova)

# Adicionando os p-values da ANOVA à tabela de estatísticas
tabela_estatisticas["p_value_anova"] = p_values_anova

# Exibindo a tabela de estatísticas com os p-values da ANOVA
display(tabela_estatisticas)
```

Método de Poisson

```python
from scipy.stats import poisson

# como a partir das medias calcular a probabilidade do time ganhar/perder/empate

def calcular_pontuacao_esperada(linha):
    time_casa = linha["casa"]
    time_fora = linha["fora"]

    # qual a média de gols que o time x, jogando em casa, faz contra o time y, jogando fora de casa
    lambda_casa = (tabela_estatisticas.loc[tabela_estatisticas["time"]==time_casa, "gols_feitos_casa"].iloc[0]
                * tabela_estatisticas.loc[tabela_estatisticas["time"]==time_fora, "gols_sofridos_fora"].iloc[0])

    # média de gols que o o time y faz fora de casa * media de gols que o time x sofre dentro de casa
    lambda_fora =  (tabela_estatisticas.loc[tabela_estatisticas["time"]==time_fora, "gols_feitos_fora"].iloc[0]
                    * tabela_estatisticas.loc[tabela_estatisticas["time"]==time_casa, "gols_sofridos_casa"].iloc[0])




    pv_casa = 0
    p_empate = 0
    pv_fora = 0

    for gols_casa in range(8):
        for gols_fora in range(8):
            probabilidade_resultado = poisson.pmf(gols_casa, lambda_casa) * poisson.pmf(gols_fora, lambda_fora)
            if gols_casa == gols_fora:
                p_empate += probabilidade_resultado
            elif gols_casa > gols_fora:
                pv_casa += probabilidade_resultado
            elif gols_casa < gols_fora:
                pv_fora += probabilidade_resultado

    print(pv_casa)
    print(p_empate)
    print(pv_fora)

    ve_casa = pv_casa * 3 + p_empate
    ve_fora = pv_fora * 3 + p_empate
    linha["pontos_casa"] = ve_casa
    linha["pontos_fora"] = ve_fora
    return linha
```

```python
tabela_jogos_faltantes = tabela_jogos_faltantes.apply(calcular_pontuacao_esperada, axis=1)
display(tabela_jogos_faltantes)
```

```python
def ajustar_nome_time(linha):
    for nome in nomes_times:
        if nome in linha["Equipe"]:
            return nome

tabela_classificacao["time"] = tabela_classificacao.apply(ajustar_nome_time, axis=1)
tabela_classificacao_atualizada = tabela_classificacao[["time", "Pts"]]
tabela_classificacao_atualizada["Pts"] = tabela_classificacao_atualizada["Pts"].replace('25[b]','25').astype(int)


display(tabela_classificacao_atualizada)
```

```python
tabela_pontuacao_casa = tabela_jogos_faltantes.groupby("casa").sum(numeric_only=True)[["pontos_casa"]]
tabela_pontuacao_fora = tabela_jogos_faltantes.groupby("fora").sum(numeric_only=True)[["pontos_fora"]]
display(tabela_pontuacao_casa)
display(tabela_pontuacao_fora)
```

```python
def atualizar_pontuacao_previsao(linha):
    time = linha["time"]
    pontuacao = int(linha["Pts"]) + float(tabela_pontuacao_casa.loc[time, "pontos_casa"]) + float(tabela_pontuacao_fora.loc[time, "pontos_fora"])
    return pontuacao


tabela_classificacao_atualizada["Pts"] = tabela_classificacao_atualizada.apply(atualizar_pontuacao_previsao, axis=1)
tabela_classificacao_atualizada = tabela_classificacao_atualizada.sort_values(by="Pts", ascending=False)
tabela_classificacao_atualizada = tabela_classificacao_atualizada.reset_index(drop=True)
tabela_classificacao_atualizada.index = tabela_classificacao_atualizada.index + 1
print("Projeção da Tabela Final")
display(tabela_classificacao_atualizada)
```

```python
from scipy.stats import poisson, norm  # Adicionando a importação de 'norm' de 'scipy.stats'
import numpy as np

# Define o nível de confiança desejado (por exemplo, 95%)
confianca = 0.95

# Calcula o tamanho da amostra (número de jogos restantes)
n = len(tabela_jogos_faltantes)

# Calcula o intervalo de confiança para cada time usando a distribuição de Poisson
for _, row in tabela_estatisticas.iterrows():
    time_casa = row['time']
    lambda_casa = row['gols_feitos_casa'] * row['gols_sofridos_fora']
    lambda_fora = row['gols_feitos_fora'] * row['gols_sofridos_casa']

    # Calcula a média e o desvio padrão da distribuição de Poisson
    media = lambda_casa + lambda_fora
    desvio_padrao = np.sqrt(media)

    # Calcula o intervalo de confiança usando o método de Poisson
    ic_inf = media - norm.ppf((1 + confianca) / 2) * desvio_padrao / np.sqrt(n)
    ic_sup = media + norm.ppf((1 + confianca) / 2) * desvio_padrao / np.sqrt(n)

    print(f"Intervalo de Confiança para {time_casa}: [{ic_inf}, {ic_sup}]")

```

```python
from scipy.stats import norm

# Define o nível de confiança desejado (por exemplo, 95%)
confianca = 0.95

# Calcula o tamanho da amostra (número de jogos restantes)
n = len(tabela_jogos_faltantes)

# Calcula o intervalo de confiança para cada time usando a distribuição de Poisson
for _, row in tabela_estatisticas.iterrows():
    time_casa = row['time']
    lambda_casa = row['gols_feitos_casa'] * row['gols_sofridos_fora']
    lambda_fora = row['gols_feitos_fora'] * row['gols_sofridos_casa']

    # Calcula a média e o desvio padrão da distribuição de Poisson
    media = lambda_casa + lambda_fora
    desvio_padrao = np.sqrt(media)

    # Calcula o intervalo de confiança usando o método de Poisson
    ic_inf = media - norm.ppf((1 + confianca) / 2) * desvio_padrao / np.sqrt(n)
    ic_sup = media + norm.ppf((1 + confianca) / 2) * desvio_padrao / np.sqrt(n)

    tabela_estatisticas.loc[tabela_estatisticas['time'] == time_casa, 'IC_inf'] = ic_inf
    tabela_estatisticas.loc[tabela_estatisticas['time'] == time_casa, 'IC_sup'] = ic_sup

# Primeiro, vamos unir a tabela de classificação final com os intervalos de confiança
tabela_final_com_ic = tabela_classificacao_atualizada.merge(tabela_estatisticas[['time', 'IC_inf', 'IC_sup']], on='time')

# Ordena a tabela final com intervalo de confiança pela pontuação (Pts) em ordem decrescente
tabela_final_com_ic = tabela_final_com_ic.sort_values(by='Pts', ascending=False)

# Reinicia o índice após a ordenação
tabela_final_com_ic = tabela_final_com_ic.reset_index(drop=True)

# Atualiza o índice para começar em 1
tabela_final_com_ic.index = tabela_final_com_ic.index + 1

# Exibe a tabela classificativa final com intervalo de confiança
print("Projeção da Tabela Final com Intervalo de Confiança:")
display(tabela_final_com_ic)
```

```python
import numpy as np

# Calcula a pontuação esperada de cada time com base na distribuição normal
tabela_estatisticas['pontos_esperados'] = (tabela_estatisticas['gols_feitos_casa'] + tabela_estatisticas['gols_feitos_fora']) * 3

# Calcula o desvio padrão da pontuação esperada de cada time
desvio_pontos = np.sqrt(tabela_estatisticas['gols_feitos_casa'] + tabela_estatisticas['gols_feitos_fora'])

# Simula os resultados dos jogos restantes
num_simulacoes = 1000
pontuacoes_simuladas = []

for _ in range(num_simulacoes):
    pontuacao_simulada = np.random.normal(tabela_estatisticas['pontos_esperados'], desvio_pontos)
    pontuacoes_simuladas.append(pontuacao_simulada)

# Calcula a média das pontuações simuladas para cada time
pontuacoes_medias = np.mean(pontuacoes_simuladas, axis=0)

# Identifica o time com a maior pontuação média como o provável campeão
indice_campeao = np.argmax(pontuacoes_medias)
time_campeao = tabela_estatisticas.iloc[indice_campeao]['time']

print("O provável campeão da temporada é:", time_campeao)

import scipy.stats as stats

# Define o nível de confiança desejado (por exemplo, 95%)
confianca = 0.95

# Calcula o Z-score correspondente ao nível de confiança
z_score = stats.norm.ppf((1 + confianca) / 2)

# Define o tamanho da amostra
n = len(tabela_estatisticas)

# Calcula o intervalo de confiança para cada time
tabela_estatisticas['IC_inf'] = tabela_estatisticas['pontos_esperados'] - z_score * desvio_pontos / np.sqrt(n)
tabela_estatisticas['IC_sup'] = tabela_estatisticas['pontos_esperados'] + z_score * desvio_pontos / np.sqrt(n)

# Exibe o intervalo de confiança para cada time
print("Intervalo de Confiança para cada Time:")
for _, row in tabela_estatisticas.iterrows():
    print(f"{row['time']}: [{row['IC_inf']}, {row['IC_sup']}]")
```

COMPARATIVO POISSON X DISTRIBUIÇÃO NORMAL

```python
import scipy.stats as stats

# Define o nível de confiança desejado (por exemplo, 95%)
confianca = 0.95

# Calcula o tamanho da amostra
n = len(tabela_classificacao_atualizada)

# Calcula o desvio padrão das pontuações esperadas para a distribuição normal
desvio_pontos_normal = np.sqrt(tabela_estatisticas['gols_feitos_casa'] + tabela_estatisticas['gols_feitos_fora'])

# Calcula o Z-score correspondente ao nível de confiança para a distribuição normal
z_score_normal = stats.norm.ppf((1 + confianca) / 2)

# Calcula o intervalo de confiança para cada time usando a distribuição normal
tabela_estatisticas['IC_inf_normal'] = tabela_estatisticas['pontos_esperados'] - z_score_normal * desvio_pontos_normal / np.sqrt(n)
tabela_estatisticas['IC_sup_normal'] = tabela_estatisticas['pontos_esperados'] + z_score_normal * desvio_pontos_normal / np.sqrt(n)

# Calcula o intervalo de confiança para cada time usando o método de Poisson
tabela_estatisticas['IC_inf_poisson'] = tabela_estatisticas['pontos_esperados'] - stats.norm.ppf((1 + confianca) / 2) * np.sqrt(tabela_estatisticas['pontos_esperados'] / n)
tabela_estatisticas['IC_sup_poisson'] = tabela_estatisticas['pontos_esperados'] + stats.norm.ppf((1 + confianca) / 2) * np.sqrt(tabela_estatisticas['pontos_esperados'] / n)

# Ordena a tabela de estatísticas de acordo com a classificação final e renumera o índice
tabela_estatisticas = tabela_estatisticas.merge(tabela_classificacao_atualizada, on="time", how="left").sort_values(by="Pts", ascending=False).reset_index(drop=True)

# Exibe o intervalo de confiança para cada time no mesmo formato da tabela de classificação final
print("Tabela Final com Intervalo de Confiança para cada Time (Distribuição Normal e Poisson - Ordenado pela Classificação Final):")
for i, row in tabela_estatisticas.iterrows():
    print(f"{i + 1}. {row['time']}: [{row['IC_inf_normal']}, {row['IC_sup_normal']}] (Normal) | [{row['IC_inf_poisson']}, {row['IC_sup_poisson']}] (Poisson)")
```

Obtenção dos Dados: Os dados da Premier League de 2023-2024 são obtidos a partir da Wikipedia por meio de web scraping usando a biblioteca Pandas.

Intervalo de Confiança (Poisson): Calcula-se o intervalo de confiança para cada time usando o método de Poisson, levando em consideração a média de pontos esperados.

Projeção da Tabela Final com Intervalo de Confiança: A tabela final é projetada, levando em conta os intervalos de confiança calculados para ambas as distribuições (Normal e Poisson).

**Poisson:**

Prós:

Modelo amplamente utilizado para prever resultados em esportes, especialmente em jogos de futebol.
Leva em consideração a média de gols feitos e sofridos por cada time.
Pode ser adaptado para prever pontuações em partidas de futebol.

**Contras:**

Assume independência entre os eventos (gols marcados em uma partida são independentes dos gols marcados em outras).
Não leva em conta fatores como a qualidade do time adversário, condições climáticas, lesões de jogadores, entre outros.
Sensível a outliers e variações extremas nos dados.
Distribuição Normal:

Prós:

Ampla aplicabilidade em diferentes contextos estatísticos.
Utiliza o conceito de média e desvio padrão para prever resultados.
Mais robusto em casos onde há uma grande quantidade de dados disponíveis.

**Contras:**

Pode não ser a melhor escolha para prever resultados em esportes, onde os eventos não seguem uma distribuição normal (por exemplo, distribuição de gols em uma partida de futebol).
Não leva em conta a natureza discreta dos eventos em jogos esportivos, como o número inteiro de gols marcados.

**Conclusão:**

Ambos os métodos têm suas vantagens e desvantagens. No entanto, para a previsão de resultados em jogos de futebol, especialmente em uma liga competitiva como a Premier League, o método de Poisson tende a ser mais apropriado. Isso ocorre porque o modelo de Poisson é especificamente projetado para lidar com eventos discretos, como gols em uma partida de futebol, e leva em consideração a média de gols feitos e sofridos por cada time. Além disso, ao ajustar os parâmetros do modelo para cada time individualmente, é possível capturar melhor as características específicas de cada equipe. Portanto, para esta situação, o método de Poisson é mais confiável e adequado para prever os resultados dos jogos na Premier League.
