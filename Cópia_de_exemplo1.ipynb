```python
import os
import tensorflow as tf
from tensorflow import keras
import numpy as np


def data_encode():
  X = []
  Y = []

  X.append([1, 1, 0, 1])
  X.append([0, 0, 1, 0])
  X.append([1, 1, 0, 0])
  X.append([1, 0, 1, 1])
  X.append([1, 0, 0, 1])
  X.append([0, 0, 1, 1])

  Y.append([1])
  Y.append([0])
  Y.append([0])
  Y.append([1])
  Y.append([0])
  Y.append([1])

  print(X)
  print(Y)

  return X, Y

learningrate = 0.1
training_epochs = 1000

n_input = 4
n_hidden = 2
n_output = 1

train_X, train_Y = data_encode()

model = keras.Sequential([
    keras.layers.Dense(n_input, input_dim=n_input, activation='sigmoid'),
    keras.layers.Dense(n_hidden, activation='sigmoid'),
    keras.layers.Dense(n_output, activation='sigmoid')
])

optimizer = keras.optimizers.Adam(learning_rate=learningrate)

model.compile(optimizer=optimizer,
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(train_X, train_Y, epochs=training_epochs)

while True:
    print('Digite os quatro sintomas')

    s1 = int(input())
    s2 = int(input())
    s3 = int(input())
    s4 = int(input())

    X_test = np.array([[s1, s2, s3, s4]])

    predictions = model(X_test)

    if predictions[0] < 0.5:
        print('Saudvel')
    else:
        print('Doente')
```
